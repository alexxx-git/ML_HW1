```python
def new_mileage(old_column):
  newcol=[]
  for row in old_column:
    if isinstance(row, str):
      if row.split()[1]=='km/kg':
        newcol.append(round(float(row.split()[0])*1.4,1))
      else:
        newcol.append(round(float(row.split()[0]),1))
    else:
      newcol.append(None)
  return newcol
```
